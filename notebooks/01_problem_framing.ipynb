```python
## Problem Context
Fraudulent online transactions represent a very small fraction of all transactions, making fraud detection an imbalanced classification problem. A naive model can achieve high accuracy by predicting most transactions as legitimate while failing to detect fraud. This project focuses on building a model that correctly identifies fraudulent transactions using evaluation metrics that reflect performance on the minority class.
```

```python
## Problem Statement
This project aims to develop a machine learning model that supports automated fraud prevention by identifying high-risk online transactions. The model is designed to automatically block transactions with a high likelihood of fraud while flagging borderline cases for manual review. The primary consumer of the modelâ€™s output is an automated fraud prevention system, with secondary use by fraud analysts who review flagged transactions. In this context, false negatives are considered more costly because undetected fraudulent transactions lead to direct financial loss, whereas false positives mainly result in temporary transaction delays and increased operational workload.
```

```python
## Decision Supported
- Automatically block transactions with a high predicted fraud risk
- Flag borderline transactions for manual review
```

```python
## Evaluation Metrics
Primary metrics (fraud class):
- Recall
- Precision
- PR-AUC

Supporting analysis:
- Confusion matrix
- Threshold tuning to balance false positives and false negatives

Accuracy is not used as a primary metric due to extreme class imbalance.
```

```python
## Assumptions and Limitations
- The system is allowed to automatically block high-risk transactions
- The dataset is anonymized and may not include all real-world fraud signals
- Fraud labels may be noisy or delayed
```

```python
## Definition of Success
A successful model improves recall and PR-AUC over a logistic regression baseline while maintaining practical precision levels, supported by confusion matrix analysis and a clearly justified decision threshold.
```
